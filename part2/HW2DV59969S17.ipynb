```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
import scipy.stats

pixels = data['B1'].shape
#print(pixels)
N = pixels[0]*pixels[1]
print('Number of pixels N:',N)

spercentage = 0.45
numSamps = round(N*spercentage)
print('Sample of 45%: ' + str(numSamps))

samplesInds = list(range(1,N))
np.random.shuffle(samplesInds)
samplesInds = samplesInds[0:numSamps]
```

# #5-#9

```python
b1 = data['B1']
b1_ = b1.ravel()[samplesInds]

b6 = data['B6']
b6_ = b6.ravel()[samplesInds]


fig = plt.figure(figsize=(12.0,12.0))
grid = mpl.gridspec.GridSpec(1,1)
ax = fig.add_subplot(grid[0])

ax.spines['top'].set_visible(False)
ax.spines['bottom'].set_linewidth(0.5)
ax.xaxis.set_ticks_position('bottom')

ax.spines['right'].set_visible(False)
ax.spines['left'].set_linewidth(0.5)
ax.yaxis.set_ticks_position('left')

ax.xaxis.set_tick_params('major', length=2, labelsize=20)
ax.yaxis.set_tick_params('major', length=2, labelsize=20)

ax.set_xlim((0.4,1.1))
ax.set_ylim((0,0.75))

ax.plot(b1_,b6_, marker='.', color='black', linestyle='',markersize=2, alpha=.1 ,clip_on=False)
ax.autoscale_view()

plt.title("Band 1 vs. Band 6 Reflectance",fontsize=20)
ax.set_xlabel('Band 1 Reflectance', fontsize=20)
ax.set_ylabel('Band 6 Reflectance', fontsize=20)

linePoly = np.polyfit(b1_,b6_,1)
y1 = np.polyval(linePoly,1.1) 
y2 = np.polyval(linePoly,0.90)


plt.plot(b1_,np.polyval(linePoly,b1_),'-',label='Linear Regression')
plt.legend(bbox_to_anchor=(1, 1), loc=9, borderaxespad=5.,fontsize=20)

slope = ('Slope = ' + str(linePoly[0]))
coeff = ('\n Pearson Coeficient:' + str(scipy.stats.pearsonr(b1_,b6_)))
ax.text(1.23, .65, slope + coeff, verticalalignment='bottom', horizontalalignment='right', fontsize=20)
plt.show()
```

```python
b7 = data['B7']
b7_ = b7.ravel()[samplesInds]

b6 = data['B6']
b6_ = b6.ravel()[samplesInds]


fig = plt.figure(figsize=(12.0,12.0))
grid = mpl.gridspec.GridSpec(1,1)
ax = fig.add_subplot(grid[0])

ax.spines['top'].set_visible(False)
ax.spines['bottom'].set_linewidth(0.5)
ax.xaxis.set_ticks_position('bottom')

ax.spines['right'].set_visible(False)
ax.spines['left'].set_linewidth(0.5)
ax.yaxis.set_ticks_position('left')

ax.xaxis.set_tick_params('major', length=2, labelsize=20)
ax.yaxis.set_tick_params('major', length=2, labelsize=20)

ax.set_xlim((0,0.5))
ax.set_ylim((0,0.6))

ax.plot(b1_,b6_, marker='.', color='black', linestyle='',markersize=2, alpha=.1 ,clip_on=False)
ax.autoscale_view()

plt.title("Band 7 vs. Band 6 Reflectance",fontsize=20)
ax.set_xlabel('Band 7 Reflectance', fontsize=20)
ax.set_ylabel('Band 6 Reflectance', fontsize=20)

linePoly = np.polyfit(b7_,b6_,1)
y1 = np.polyval(linePoly,1.1) 
y2 = np.polyval(linePoly,0.90)


plt.plot(b7_,np.polyval(linePoly,b7_),'-',label='Linear Regression')
plt.legend(bbox_to_anchor=(1, 1), loc=9, borderaxespad=5.,fontsize=20)

slope = ('Slope = ' + str(linePoly[0]))
coeff = ('\n Pearson Coeficient:' + str(scipy.stats.pearsonr(b7_,b6_)))
ax.text(0.77, .65, slope + coeff, verticalalignment='bottom', horizontalalignment='right', fontsize=20)
plt.show()
```
